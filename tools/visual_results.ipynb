```python
import numpy as np, os, collections, matplotlib.pyplot as plt
results_dir = '../results/'
results_sub_dir = os.listdir(results_dir)
results_sub_dir.sort()
table = collections.defaultdict(list)

for item in results_sub_dir:
    loss = np.load(os.path.join(results_dir, item, 'sinkhorn_div.npy'))
    name = item[:item.find('S[')]
    table[name].append(loss)

print(len(table.keys()), table.keys())
```

```python
for key in table.keys():
    table[key] = np.array(table[key])
    table[key] = table[key].mean(0)
    # print(table[key].shape)
```

```python
plt.figure(figsize = (19.2, 10.8))
for key in table.keys():
    plt.plot(table[key], label = key)
plt.legend()
plt.yscale('log')
plt.ylim([2.3e-4, 3.4e-4])
plt.show()
```
